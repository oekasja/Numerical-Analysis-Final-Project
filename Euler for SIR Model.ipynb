```python
import numpy as np

# Parameter
S = 8377828
I = 8000
R = 939
N = 8386763
A = 0.000046
B = 0.326666
C = 0.123111
D = 0.041230
E = 0.003700
x0 = 0.998935
y0 = 0.000954
z0 = 0.000112
h = 0.1
n_iterasi = 100

# Buat array untuk menyimpan hasil
x = np.zeros(n_iterasi + 1)
y = np.zeros(n_iterasi + 1)
z = np.zeros(n_iterasi + 1)

# Set nilai awal
x[0] = x0
y[0] = y0
z[0] = z0

# Iterasi menggunakan metode Euler
for i in range(n_iterasi):
   x[i + 1] = x[i] + h * (A - (B * x[i]) - (C * B * x[i] * y[i]) - (A * x[i]))
   y[i + 1] = y[i] + h * ((B * x[i]) - (A + D) * y[i])
   z[i + 1] = z[i] + h * ((C * B * x[i] * y[i]) - (A + E) * z[i])

# Print hasil
print("x:", x)
print("y:", y)
print("z:", z)
```

```python
import matplotlib.pyplot as plt

# Buat plot
plt.plot(x, label="x = Populasi yang Rentan")
plt.plot(y, label="y = Populasi yang Terinfeksi I")
plt.plot(z, label="z = Populasi yang Terinfeksi II")
plt.legend()
plt.xlabel("Iterasi/Waktu")
plt.ylabel("Nilai")

# Tampilkan plot
plt.show()
```
